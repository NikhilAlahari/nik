```python
import cv2                          #Importing cv2 module
import numpy as np                  #Importing the cv2
import matplotlib.pyplot as plt     #Importing the matplot as plt
%matplotlib inline
```

```python
bw = cv2.imread("detect_blob.png",0)  #Loading the Image
plt.imshow(bw)                        #Plotting the Image
```

```python
height, width = bw.shape[0:2]                #getting the height and width
binary = np.zeros([height,width,1],'uint')   
thresh = 100                                 #Setting up the threshold to 100
```

```python
for row in range(0,height):            #first for going through heights
    for col in range(0,width):         #second for loop for going through width
        if bw[row][col]>thresh:        #if more than threshold we will not loose any data and not eliminate it
            binary[row][col] = 255     #if more threshold then we will turn it off as its less intensity
plt.imshow(binary)                     #plotting the image
```

Contours -- boundaried around objects in an image

```python
img = cv2.imread('detect_blob.png',1)
gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
thresh = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
plt.imshow(thresh)
```

```python
contours,hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE) #

img2 = img.copy()
index= -1 #All Objects
thickness = 4
color = (255,0,255)

cv2.drawContours(img2,contours,index,color,thickness)
plt.imshow(img2)
```

```python
# Faces
img = cv2.imread("faces.jpeg",1)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
path = "haarcascade_frontalface_default.xml"
```

```python
face_cascade = cv2.CascadeClassifier(path)
```

```python
faces = face_cascade.detectMultiScale(gray,scaleFactor=1.055,minNeighbors = 5,minSize=(40,40))
```

```python
print(len(faces))
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
```
